Considering,all test scores are to be 100.

===>finding percentage

```python
schl_per['total score']=schl_per['writing score']+schl_per['reading score']+schl_per['math score']
schl_per['percentage']=schl_per['total score']/300*100
schl_per
```

considering grades, for the test scores 
where

GRADE O=90 TO 100 percentage.

GRADE A+=80 TO 90 percentage.

GRADE A=70 TO 80 percentage.

GRADE B+=60 TO 70 percentage.

GRADE B=50 TO 60 percentage. 

GRADE C+=40 TO 50 percentage.

GRADE C=30 TO 40 percentage 
 
 FAIL =0 TO 30 percentage 

```python
def grade(perc):
    if perc>= 90 and perc<= 100:
        return 'Grade O'
    elif perc>= 80and perc<90:
        return 'Grade A+'
    elif perc>= 70 and perc<80:
        return 'Grade A'
    elif perc>= 60and perc<70:
        return 'Grade B+'
    elif perc>= 50and perc<60:
        return 'Grade B'
    elif perc>= 40and perc<50:
        return 'Grade C+'
    elif perc>= 30 and perc<40:
        return 'Grade C'
    else:
        return 'FAIL'
    
schl_per['grades']=schl_per['percentage'].apply(grade)
```
